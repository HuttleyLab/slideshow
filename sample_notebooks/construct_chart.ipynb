Mermaid diagrams can be generated directly in markdown

```mermaid
graph LR
  s((start))
  e((end))
  s-->CAC-->ACA-->CAG-->AGC-->GCA-->CAT-->e
  CAC-->ACG-->CGA-->GAG-->AGC
```

````python
# Or generated from code using nbtools.mm(graph: str)
from nbtools import mm 
mm('''
   graph LR
   s((start)) 
   e((end))
   s-->CAC-->ACA-->CAG-->AGC-->GCA-->CAT-->e  
   CAC-->ACG-->CGA-->GAG-->AGC''')
````

````python
# Graphviz diagrams can also be rendered in cell results
from nbtools import gv
gv("""
graph {
    a -- b;
    b -- c;
    a -- c;
    d -- c;
}
""")
````
